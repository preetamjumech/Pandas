### Exercise 9: Concatenate two data frames using the following conditions

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
german_df = pd.DataFrame(GermanCars)
japaneese_df = pd.DataFrame(japaneseCars)
concat_df = pd.concat([german_df,japaneese_df],keys=["German","Japan"])
concat_df
```

```python
german_df1 = pd.DataFrame.from_dict(GermanCars)
german_df1.head()
```

### Exercise 10: Merge two data frames using the following condition

```python
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
car_price_df = pd.DataFrame.from_dict(Car_Price)
car_horsepower_df = pd.DataFrame.from_dict(car_Horsepower)
merge_df = pd.merge(car_price_df,car_horsepower_df,on="Company")
merge_df
```
