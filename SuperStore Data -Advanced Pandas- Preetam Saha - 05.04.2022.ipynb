### Exercise 8: Missing values in Pandas

```python
df_missing = pd.read_excel("Sample - Superstore.xls",sheet_name="Missing")
df_missing.head()
```

```python
df_missing.isnull().sum()
```

```python
df_missing.isnull()
```

```python
for c in df_missing.columns:
    miss = df_missing[c].isnull().sum()
    if miss>0:
        print(f"{c} has {miss} missing value(s)")
    else:
        print(f"({c} has no missing values!!!")
```

### Exercise 13: Merging by a common key

### Exercise 19: Flexibility for user-defined function with apply method

```python
df.head()
```

```python
def category(price):
    if price<50:
        return "Low"
    elif price <200:
        return "Medium"
    else:
        return "High"
```

```python
df_sample.head()
```

```python
df_sample["Category"]=df_sample["Sales"].apply(category)
df_sample.head()
```

```python
df_sample["Customer_name_length"]=df_sample["Customer Name"].apply(len)
df_sample.head()
```

```python
df_sample["Price_after_By_default_Discount"]=df_sample["Sales"].apply(lambda x: x*0.85 if x>0 else x)
df_sample.head()
```
